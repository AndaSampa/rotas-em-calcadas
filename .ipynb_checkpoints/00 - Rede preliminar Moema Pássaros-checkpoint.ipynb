# Rede caminhável preliminar de Moema Pássaros

Neste ensaio inicial vamos utilizar Python para geoprocessar os dados de uma região homogênia e reticulada do distrito de Moema para gerar um conhecimento básico inicial e podermos pensar sobre o processamento em regiões com morfologia mais heterogênea e orgânica

## Obtendo o ponto do lote para conectar a rede

Primeiramente vamos obter um ponto no lote próximo a quadra predial para poder conecta-lo a rede. O ideal nesse caso seria utilizar a numeração de porta e poder usar esse processamento para geocidificação de endereços. No entanto.

O processamento é basicamente o recorte de um buffer de 1 metro para dentro do lote, processar a diferença booleana e a partir do polígono resultante obter o ponto mais representativo. Que é, a grosso modo, um centroíde com a garantia que esteja contido na geometria.

Consideramos aqui que o ponto deva estar contido na geometria do lote para realizar operações espaciais.

```python
import geopandas as gpd
```

```python
# Carregando lotes
gdf_lotes = gpd.read_file('gis-data/moema-passaros.gpkg', layer='lotes')
```

```python
# Carregando Quadras
gdf_quadras_prediais = gpd.read_file('gis-data/moema-passaros.gpkg', layer='quadras-prediais')
```

```python
gdf_quadras_prediais.buffer(-1).plot()
```

```python
# Processando buffer, operação de diferença e ponto representativo em uma linha
gdf_lote_porta = \
gdf_lotes.difference(gdf_quadras_prediais.buffer(-1).unary_union).representative_point()
```

```python
gdf_lote_porta.plot()
```

```python
# Salvando o processamento 
gdf_lote_porta.to_file('resultados/moema-passaros-lote-porta.gpkg', driver='GPKG')
```

## Conectar o ponto do lote à rede

Agora que temos um ponto para cada lote podemos começar a montar a rede, criando uma linah de conexão entre o ponto e o eixo medial da calçada que foi previamente processado em PostGis

```python
from shapely.ops import nearest_points
from shapely.geometry import LineString, MultiLineString, Point
```

```python
gdf_eixos_mediais = gpd.read_file('gis-data/moema-passaros.gpkg', layer='eixos-mediais')
```

```python
## TODO melhorar o desempenho
# Para cada ponto de lote que não esteja vazio é criado uma linha com a menor distância até o eixo medial da calçada
gdf_ponto_lote_rede = \
gdf_lote_porta[~gdf_lote_porta.is_empty].apply(lambda x: nearest_points(x, gdf_eixos_mediais.unary_union))
```

```python
gpd.GeoDataFrame(geometry=gdf_ponto_lote_rede.apply(lambda x: LineString(x))).plot()
```

```python
# Cria e salva o segmento de rede que conecta o ponto à rede em uma camada chamada 'segmento-de-rede'
gpd.GeoDataFrame(geometry=gdf_ponto_lote_rede.apply(lambda x: LineString(x)))\
.to_file('resultados/moema-passaros-lote-porta.gpkg', layer='segmento-de-rede')
```

## Conectando os eixos mediais com travesia dos logradouros

Agora que temos as conexões entre o eixo medial e os lotes é necessário conectar cada rede preliminar de entre as quadaras. 

O ideal para esse processamento seria identificar as faixas de pedestre e elementos de travassia, como passagens subterrâneas, passarelas e viadutos. No entanto, de forma provisória, vamos utilizar o polígono de vias para inferir a faixa de pedestre.

```python
from shapely.ops import linemerge
```

```python
gdf_poligono_de_via = \
gpd.read_file('gis-data/moema-passaros.gpkg', layer='poligono-de-vias')
```

```python
gdf_quadras_viarias = \
gpd.read_file('gis-data/moema-passaros.gpkg', layer='quadras-viarias')
```

```python
# cria uma abstração do 'meio fio' um elemento geométrico em linha para cada face de quadra 
gdf_meio_fio = gdf_poligono_de_via.intersection(gdf_quadras_viarias.buffer(0.5).boundary.unary_union)
```

```python
# Conecta cada um dos lados do meio-fio nos seus pontos mais extremos e
# conecta essas mesmos pontos ao ponto mais próximo do eixo medial das calçadas

def connect_sides(x):
    lm = linemerge(x)
    eixos = gdf_eixos_mediais.unary_union
    
    if lm.geom_type == 'MultiLineString':
#         print(list(lm.geoms)[0].coords[:][-1])
        return MultiLineString([LineString([list(lm.geoms)[0].coords[:][0], list(lm.geoms)[1].coords[:][-1]]),\
                               LineString([list(lm.geoms)[1].coords[:][0], list(lm.geoms)[0].coords[:][-1]]),\
                               LineString(nearest_points(Point(list(lm.geoms)[0].coords[:][0]), eixos)),\
                                LineString(nearest_points(Point(list(lm.geoms)[0].coords[:][-1]), eixos)),\
                                LineString(nearest_points(Point(list(lm.geoms)[1].coords[:][-1]), eixos)),\
                               LineString(nearest_points(Point(list(lm.geoms)[1].coords[:][0]), eixos))])
#         return LineString(nearest_points(Point(list(lm.geoms)[0].coords[:][0]), eixos))
    return None

gdf_meio_fio.apply(lambda x: connect_sides(x)).plot()
```

```python
# Salva as linhas em uma camadas chamada 'travessia'
gdf_meio_fio.apply(lambda x: connect_sides(x)).\
to_file('resultados/moema-passaros-lote-porta.gpkg', \
        layer='travessias', driver='GPKG')
```
